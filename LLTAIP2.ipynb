```python
# IMPORT REQUIRED LIBRARIES
```

```python
from deep_translator import GoogleTranslator
```

```python
# CREATE LANGUAGE DICTIONARY
```

```python
languages = {
    "English": "en",
    "Hindi": "hi",
    "French": "fr",
    "Spanish": "es",
    "German": "de",
    "Chinese": "zh-CN",
    "Japanese": "ja",
    "Korean": "ko",
    "Punjabi": "pa",
    "Italian": "it",
    "Russian": "ru",
    "Arabic": "ar"
}
```

```python
# DISPLAY AVAILABLE LANGUAGES
```

```python
print("Available Languages:\n")

for language in languages:
    print(language)
```

```python
# TAKE USER INPUT
```

```python
source_language = input("Enter Source Language: ").title()
target_language = input("Enter Target Language: ").title()
text = input("Enter Text to Translate: ")
```

```python
translator = GoogleTranslator(
    source=languages[source_language],
    target=languages[target_language]
)

translated_text = translator.translate(text)
```

```python
#SHOW TRANSLATED TEXT
```

```python
print("\nOriginal Text:")
print(text)

print("\nTranslated Text:")
print(translated_text)
```

```python
# TRANSLATE MULTIPLE SENTENCES
```

```python
while True:
    source_language = input("\nEnter Source Language: ").title()
    target_language = input("Enter Target Language: ").title()
    text = input("Enter Text to Translate: ")

    translator = GoogleTranslator(
        source=languages[source_language],
        target=languages[target_language]
    )

    translated_text = translator.translate(text)

    print("\nTranslated Text:")
    print(translated_text)

    choice = input("\nDo you want to translate another sentence? (yes/no): ").lower()

    if choice != "yes":
        print("\nThank you for using the Language Translation Tool!")
        break
```

```python
# SAVE LANGUAGE CONFIGURATION
```

```python
import pickle

with open("languages.pkl", "wb") as file:
    pickle.dump(languages, file)

print("Language configuration saved successfully!")
```

```python
# LOAD LANGUAGE CONFIGURATION
```

```python
import pickle

with open("languages.pkl", "rb") as file:
    loaded_languages = pickle.load(file)

print("Language configuration loaded successfully!")
print(loaded_languages)
```
